Pergunta 1: Você teria interesse em fazer cursos gratuitos de tecnologia pelo celular?

```python
import matplotlib.pyplot as plt

# Gráfico 1: Interesse em cursos gratuitos
interesse_labels = ['Sim', 'Não', 'Talvez']
interesse_values = [32, 3, 5]

plt.figure(figsize=(6, 6))
plt.pie(interesse_values, labels=interesse_labels, autopct='%1.1f%%', startangle=90,
        colors=['#4CAF50', '#F44336', '#FF9800'])
plt.title('Você teria interesse em fazer cursos gratuitos de tecnologia pelo celular?')
plt.axis('equal')
plt.show()
```

Pergunta 3: Qual seu nível de experiência com tecnologia?

```python
import matplotlib.pyplot as plt

# Gráfico 3: Nível de experiência
experiencia_labels = ['Nenhuma', 'Básica', 'Intermediária', 'Avançada']
experiencia_values = [6, 20, 12, 2]

plt.figure(figsize=(6, 6))
plt.pie(experiencia_values, labels=experiencia_labels, autopct='%1.1f%%', startangle=90,
        colors=['#FE3C3A', '#026ccc', '#FFC107', '#5ebe44'])
plt.title('Qual seu nível de experiência com tecnologia?')
plt.axis('equal')
plt.show()
```
