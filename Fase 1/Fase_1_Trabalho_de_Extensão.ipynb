Pergunta 1: VocÃª teria interesse em fazer cursos gratuitos de tecnologia pelo celular?

```python
import matplotlib.pyplot as plt

# GrÃ¡fico 1: Interesse em cursos gratuitos
interesse_labels = ['Sim', 'NÃ£o', 'Talvez']
interesse_values = [32, 3, 5]

plt.figure(figsize=(6, 6))
plt.pie(interesse_values, labels=interesse_labels, autopct='%1.1f%%', startangle=90,
        colors=['#4CAF50', '#F44336', '#FF9800'])
plt.title('VocÃª teria interesse em fazer cursos gratuitos de tecnologia pelo celular?')
plt.axis('equal')
plt.show()
```
